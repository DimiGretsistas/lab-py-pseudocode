### Exercise 3: Sum of First N Natural Numbers

Write pseudocode to calculate the sum of the first N natural numbers. Then, write the equivalent Python code.

FUNCTION sum_natural_numbers(n)
    total = 0

    FOR i = 1 TO n
        total = total + i
    END FOR

    RETURN total
END FUNCTION

READ n
result = sum_natural_numbers(n)
PRINT result

```python
def sum_natural_numbers(n):
    total = 0

    for i in range(1, n + 1):
        total += i
    return total

n = int(input("Please insert a number"))
result = sum_natural_numbers(n)
print(result)
```

FUNCTION find_max(a, b, c)

### Exercise 5: Count Vowels in a String

Write pseudocode to count the number of vowels in a given string and print the count. Then, write the equivalent Python code.

FUNCTION count_vowels(text)
    count = 0

    FOR each character in text
        IF character is a vowel
            count = count + 1
        END IF
    END FOR

    RETURN count
END FUNCTION

READ text
result = count_vowels(text)
PRINT result

```python
def count_vowels(text):
    count = 0
    for char in text.lower():
        if char in "aieou":
            count += 1
    return count

text = "this is a sentence for checking the function"
count_vowels(text)
```

### Bonus Exercise 9: Find Factorial of a Number

Write pseudocode that calculates the factorial of a given positive integer.

```python
def factorial(n):
    result = 1

    for i in range(1, n + 1):
        result *= i
    return result


n = 78
answer = factorial(n)

print("Factorial:", answer)
```
